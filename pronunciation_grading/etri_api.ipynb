```python
import requests
import base64

# ETRI API URL 및 발급받은 access_key
API_URL = "http://aiopen.etri.re.kr:8000/WiseASR/Pronunciation"
ACCESS_KEY = "aab76c95-cd54-4621-9e09-e8b62467ccb7"

# 파일 읽기 (audio 파일은 WAV 형식이어야 함)
def load_audio_file(file_path):
    with open(file_path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf8")

# 발음평가 요청
def evaluate_pronunciation(audio_path, script):
    audio_data = load_audio_file(audio_path)
    
    request_data = {
        "access_key": ACCESS_KEY,
        "argument": {
            "language_code": "en",
            "script": script,
            "audio": audio_data
        }
    }

    response = requests.post(API_URL, json=request_data)
    return response.json()

# 실행 예제
if __name__ == "__main__":
    audio_file = "./bark_generation.wav"  # 분석할 음성 파일 경로
    text = "Hello, how are you?"    # 발음 평가 기준 텍스트
    
    result = evaluate_pronunciation(audio_file, text)
    print("발음 점수:", result.get("return_object", {}).get("score"))
    print("피드백:", result.get("return_object", {}).get("feedback"))
```
